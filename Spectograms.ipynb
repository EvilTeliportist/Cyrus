```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import os
import numpy as np

for filename in os.listdir('buffers/'):
    if filename.endswith(".wav"):
        print(filename)
        sample_rate, samples = wavfile.read('buffers/' + filename)
        frequencies, times, spectrogram = signal.spectrogram(samples, 16000)

        plt.pcolormesh(times, frequencies, spectrogram)
        plt.imshow(np.log(spectrogram))
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.show()
```
