```python
 # Neural network imports
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense

# Other imports
import yfinance as yf
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
data = pd.read_csv('sample_datasets/school_grades_dataset.csv')
print(data)
```

```python
data['sex'] = data['sex'].map({'M': 0, 'F':0})
```

```python
# Create training and testing data
training_inputs = data.loc[:350, ['sex', 'health', 'age']]
testing_inputs = data.loc[350:, ['sex', 'health', 'age']]
training_outputs = data.iloc[:350, 32]
testing_outputs = data.iloc[350:, 32]

# Normalize outputs
training_outputs = training_outputs / max(training_outputs)
testing_outputs = testing_outputs / max(testing_outputs)

print(training_inputs)
print(training_outputs)
```

```python
# Instantiate the model
model = Sequential()
model.add(Dense(32, input_dim=3, activation='sigmoid'))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(16, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

# Compilation
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])

# Training
EPOCHS = 500
BATCH_SIZE = 10
fit = model.fit(training_inputs[:-1], training_outputs, epochs = EPOCHS, batch_size = BATCH_SIZE)
```

```python
model.evaluate(testing_inputs, testing_outputs)
```

```python
plt.plot(fit.history['loss'])
```

```python
plt.plot(fit.history['accuracy'])
```

```python
vals = []
results = []
for i in range(100):
    sex = random.randint(0, 10) % 2
    health = random.randint(0, 5)
    age = random.randint(15, 22)
    vals.append([sex, health, age])
    results.append(model.predict([[sex, age, health]]))
```
